This question requires students to combine multiple criteria at once: filtering for planets with a radius near 1.0 radii, equilibrium temperature near 255 kelvin, distance from the host star near 1.0 AU, and a host star temperature near 5,778 kelvin. These missing values (NaNs) have been left in intentionally. Mentors should encourage students to identify and address them as part of their data cleaning process (using df.dropNa), rather than treating them as errors in the dataset.
